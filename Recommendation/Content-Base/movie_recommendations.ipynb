```python
user_input = {
    'MovieName': ['Toy Story', 'Grumpier Old Men', 'Grand Slam'],
    'Rating': [5, 3.2, 3]
}
movie_input = pd.DataFrame(user_input)
```

```python
movie_id = movies[movies['title'].isin(movie_input['MovieName'])]
movie_input = pd.merge(movie_id, movie_input, left_on='title', right_on='MovieName')
movie_input = movie_input.drop('MovieName', axis=1)
movie_input = movie_input.reset_index(drop=True)
# movie_input.head()
```

```python
user_genres_table = movie_input.drop('title', axis=1).drop('genres', axis=1).drop('year', axis=1) \
.drop('movieId', axis=1).drop('Rating', axis=1)
```

```python
weight_table = user_genres_table.transpose().dot(movie_input.Rating)
```

```python
genres_table = movies.set_index(movies.movieId)
genres_table = genres_table.drop('title', axis=1).drop('genres', axis=1).drop('movieId', axis=1).drop('year', axis=1)
```

```python
recommend_system = ((genres_table*weight_table).sum(axis=1)/ weight_table.sum())
recommend_system = recommend_system.sort_values(ascending=False)
```

```python
recommend = movies.loc[movies.movieId.isin(recommend_system.head(20).keys())].reset_index()
recommend
```
